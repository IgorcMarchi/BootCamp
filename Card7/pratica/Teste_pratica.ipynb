```python
jobs = ecom['Job'].value_counts().head(5) 
```

```python
# Grfico de barras
plt.figure(figsize=(8,5))
jobs.plot(kind='bar', color='skyblue', edgecolor='black')

plt.title("Top 5 Profisses no Dataset")
plt.xlabel("Profisso")
plt.ylabel("Quantidade")
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```
