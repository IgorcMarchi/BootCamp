```python
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
sns.set()
```

```python
x = np.linspace(0, 18, 100) # Cria um array x com 100 valores igualmente espaçados entre 0 e 18.
y = np.sin(x) # Calcula o seno de cada valor em x e armazena em y
plt.plot(x, y, color = "green", linestyle="-") # Plota o gráfico de y versus x, na cor verde e em linha continua.
plt.show() # exibe o gráfico
```

```python
w = np.linspace(0, 14, 144) # cria um array com 144 valores igualmente espaçados entre 0 a 14
t = np.sin(w) # Calcula o seno de cada valor no array w e armazena em t.
plt.plot(w, t, color = "red", linestyle="--") # Plota o gráfico de y versus x, na cor vermelha e em linha tracejada.
plt.show() # exibe o gráfico
```

```python
classes = ["A", "B", "C", "D", "E"] # define os nomes da horizonal
valores = [10, 5, 3, 5, 7] # define os valores da altura
plt.bar(classes, valores, color="blue") # define a cor das barras como azul e cria um gráfico de barras vertical
plt.show() # exibi
```

```python
x = np.random.rand(300) # 300 numeros aleatorios entre 0 e 1
y = np.random.rand(300) # 300 numeros aleatorios entre 0 e 1
plt.scatter(x, y, color="green", alpha= 0.7, edgecolor= "red") # Cria o gráfico de dispersão: cada ponto é um par, a cor dentro é verde e o contorno vermelho
plt.show # exibi
```

```python
fatias = [25, 41, 14, 20] # valores
labels = ["Brasil", "Argentina", "Uruguai", "Colômbia"] # nomes de separação

plt.pie(fatias, labels=labels, autopct="%1.1f%%", startangle=90, colors=["#ff9999","#66b3ff","#99ff99","#ffcc99"]) # cria uma grafico de pizza que mostra o valor com porcentagem e cor definida
plt.title("Chance de próximo campeão da américa") # titulo
plt.show() # exibi
```

Moda, mediana e media


```python
from scipy import stats
Z = np.random.normal(10, 20, 500) # valores random
print(np.mean(Z)) # media
print(np.median(Z)) # mediana
```

```python
idades = np.random.randint(18, high=25, size= 1000) # cria 1000 idades entre 18 e 25 de forma aleatoria
stats.mode(idades) # pega a moda das idades
```

Variância e desvio padrão


```python
vp = np.random.normal(1000.0, 250.00, 1000)
print(np.std(vp)) # variancia
print(np.var(vp)) # desvio padrão
```

Covariancia e correlação

```python
x = np.random.normal(0, 1, 100) # 100 valores random entre 0 e 1 
y = np.random.normal(0, 1, 100)

media_X = np.mean(x) # media
media_Y = np.mean(y)

covariancia = np.sum((x - media_X) * (y - media_Y)) / (len(x) - 1) # calcula a covariancia utilizando a formula
print("Covariância: ", covariancia) # mostra o valor
```

Porcentagem e momentos

```python
vals = np.random.normal(0, 2.0, 10000) # pega 10000 numeros aleatórios (array)

plt.hist(vals, 50) # cria um histograma de 50 barras
plt.show() # exibe
```

```python
np.percentile(vals, 50) # calcula o a porcentagem de 50% do conjunto de dado
```

```python
np.percentile(vals, 10) # calcula o a porcentagem de 10% do conjunto de dado
```

```python
np.percentile(vals, 99) # calcula o a porcentagem de 99% do conjunto de dado
```

```python
np.mean(vals) # media
```

```python
np.var(vals) # Variancia elevado a 2
```

```python
import scipy.stats as sp
sp.skew(vals) # assimetria
```
