Moda, mediana e media


```python
from scipy import stats
Z = np.random.normal(10, 20, 500)
print(np.mean(Z))
print(np.median(Z))
```

```python
idades = np.random.randint(18, high=25, size= 1000)
stats.mode(idades)
```

Variância e desvio padrão


```python
vp = np.random.normal(1000.0, 250.00, 1000)
print(np.std(vp))
print(np.var(vp))
```

Covariancia e correlação

```python
x = np.random.normal(0, 1, 100)
y = np.random.normal(0, 1, 100)

media_X = np.mean(x)
media_Y = np.mean(y)

covariancia = np.sum((x - media_X) * (y - media_Y)) / (len(x) - 1)
print("Covariância: ", covariancia)
```
